## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
p_white=10/(10+12+8)
print(f"Prbability of picking a white ball: {p_white:.4f}")
p_white1_black2_without_replacement = (10/30) * (8/29)
print(f"Probability of picking a white ball first and then a black ball without replacement: {p_white1_black2_without_replacement:.4f}")
p_white1_black2_with_replacement=(10/30) * (8/30)
print(f"Probability of picking a white ball first and then a black ball with replacement: {p_white1_black2_with_replacement:.4f}")

import math as m

p_3different_colors = (m.comb(10,1) * m.comb(12,1) * m.comb(8,1)) / m.comb(30,3)
print(f"Probability of picking 3 different colors without replacement: {p_3different_colors:.4f}")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
p_cloudy = 0.4
p_rain_given_cloudy = 0.5
p_rain_given_not_cloudy = 0.2

p_rain=(p_cloudy*p_rain_given_cloudy) + ((1-p_cloudy)*p_rain_given_not_cloudy)

print(f"The chance of rain during next Sarturday is: {p_rain:.2f}")
```
